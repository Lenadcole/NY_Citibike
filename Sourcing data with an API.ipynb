### 1. Import libraries and data

```python
# import libraries

import pandas as pd
import numpy as np
import os
import requests
import json
from datetime import datetime
```

```python
# create a list with all the files in the folder using a list comprehension

folderpath = r'Data'
filepaths = [os.path.join(folderpath, name) for name in os.listdir(folderpath)]
```

```python
filepaths
```

```python
# read and concatenate all files at the same time

df = pd.concat((pd.read_csv(f) for f in filepaths), ignore_index = True)
```

### 2. Explaination of code

There were two steps to reading in and joining the data.

In the first, I created a list comprehension using files located in a file named 'Data'. I did this by joining the folderpath (the 'Data' folder) and name together. This made a file path. The next part of the code then looped through each name (file) in the list that was returned from the folderpath. In simple terms, the code made a path to the correct folder and returned the list of files within.

In the second step, the code used a generator to iterate through each of the files, one at a time. It then passed the files that had been read to the pd.concat() function, which concatenated them (joined the data vertically, which was possible because each file used the same column structure and therefore did not need a key), creating the dataframe.

### 3. Obtain weather data from NOAA's web app for New York LaGuardia's weather station for 2022

```python
# define NOAA token

Token = 'XvdnslpFTxcvebTSfObYAatadEMEsazs'
```

```python
# get the API

r = requests.get('https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GHCND&datatypeid=TAVG&limit=1000&stationid=GHCND:USW00014732&startdate=2022-01-01&enddate=2022-12-31', headers={'token':Token})
```

```python
# load the API response as a json

d = json.loads(r.text)
```

```python
d
```

```python
# secure all items in response that correspond to TAVG

avg_temp = [item for item in d['results'] if item ['datatype']=='TAVG']
```

```python
# get date field from all average temperatures

dates_temp = [item['date'] for item in avg_temp]
```

```python
# get teperature from all average temperature readings

temps = [item['value'] for item in avg_temp]
```

```python
temps
```

```python
# put results in dataframe

df_temp = pd.DataFrame()
```

```python
# get the date only and cast it to date time, and convert temperature to normal Celsius from tenths of Celsius

df_temp['date'] = [datetime.strptime(d, '%Y-%m-%dT%H:%M:%S') for d in dates_temp]
df_temp['avg_Temp'] = [float(v)/10.0 for v in temps]
```
